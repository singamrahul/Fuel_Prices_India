DATA CLEANING

```python
data = data.rename(columns = {'Unnamed: 0' : 'place'})
data.sample(2)
```

```python
data['petrol'] = data['petrol'].str[:-4]
data['diesel'] = data['diesel'].str[:-4]
data
```

```python
data['diesel'] = data['diesel'].astype('float')
data['petrol'] = data['petrol'].astype('float')
data
```

```python
data.info()
```

```python
highest = data.sort_values(by = ['diesel', 'petrol'], ascending=False)
highest
```
